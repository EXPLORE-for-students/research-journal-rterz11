# 2: What? (What happened?)
## Describe what happened during your activities for the week.

In the reference material, there are two main contributions to the gravitational wave signal that are investigated. These are the sound wave and bubble wall contributions. The sound wave contributions were looked at in the previous journal. This one will focus on the bubble wall contributions. These generated a gravitational wave signal from collisions between expanding vacuum bubbles in a first order phase transition. In the thesis, the bubble wall contribution is stated in equation 4.32,

\begin{equation*}
h^{2}\Omega_{bw}(f)=\mathcal{R}h^{2}\tilde{\Omega}_{bw}(\frac{\kappa_{\phi}\alpha}{\alpha+1})^{2}(\frac{\beta}{H})^{-2}s_{bw}(f/f_{p,bw})
\end{equation*}

Where,

\begin{equation*}
s_{bw}(x)=\frac{3.8x^{2.8}}{1+2.8x^{3.8}}
\end{equation*}

These are the parameters used

    parameters = {
        'log_alpha'   : prior("Uniform", -4, 0),
        'log_beta_h'  : prior("Uniform", 0, 3),
        'log_Tp_GeV'      : prior("Uniform", -4, -1),
        'log_eps_p'   : prior("Uniform", -3, 1),
    }

The same spectrum code is used as in the previous journal. However, this time the sound wave contribution is set to 0.
